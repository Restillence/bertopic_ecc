```python
# Step 1: Import necessary modules and classes
import os
import json
import numpy as np
from bertopic import BERTopic
from sentence_transformers import SentenceTransformer
from file_handling import FileHandler
from text_processing import TextProcessor

# Step 2: Load configuration
config_file_path = r'C:\Users\nikla\OneDrive\Dokumente\winfoMaster\Masterarbeit\bertopic_ecc\config.json'
with open(config_file_path, 'r') as f:
    config = json.load(f)

# Step 3: Extract variables from configuration
index_file_ecc_folder = config["index_file_ecc_folder"]
folderpath_ecc = config["folderpath_ecc"]
sample_size = config["sample_size"]
document_split = config["document_split"]
section_to_analyze = config["section_to_analyze"]
max_documents = config["max_documents"]
model_save_path = config["model_save_path"]

# Step 4: Initialize FileHandler and TextProcessor
file_handler = FileHandler(
    index_file_path=config["index_file_path"],
    folderpath_ecc=folderpath_ecc
)
text_processor = TextProcessor(
    method=document_split,
    section_to_analyze=section_to_analyze
)

# Step 5: Read the index file and create ECC sample
index_file = file_handler.read_index_file()
ecc_sample = file_handler.create_ecc_sample(sample_size)

# Step 6: Extract all relevant sections (documents)
documents = text_processor.extract_all_relevant_sections(ecc_sample, max_documents)

# Step 7: Verify documents
print(f"Number of documents: {len(documents)}\n")
for i, doc in enumerate(documents[:5], 1):
    print(f"Document {i}:\n{doc}\n{'-'*80}\n")

# Step 8: Load the embedding model
embedding_model = SentenceTransformer("all-MiniLM-L12-v2")


```

```python
# Step 9: Load BERTopic model with the embedding model
topic_model = BERTopic.load(model_save_path, embedding_model=embedding_model)
topic_model.reduce_topics(nr_topics=20,docs=documents)
```

```python
# Step 9: Load BERTopic model with the embedding model
topic_model = BERTopic.load(model_save_path, embedding_model=embedding_model)

# Step 10: Transform documents
topics, probabilities = topic_model.transform(documents)

# Step 11: Generate visualizations
topic_model.visualize_topics()
```
